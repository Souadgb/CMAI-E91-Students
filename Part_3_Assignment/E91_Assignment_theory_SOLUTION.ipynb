## Part 1

### Question 1

If Eve intercepts and measures **100%** of the Bell pairs, what CHSH value would you expect?

A) $|S|\approx 2\sqrt{2}$

B) $|S|\le 2$

C) $|S|\approx 2.5$

**Correct:** B

If Eve measures 100% of the pairs, the CHSH value would be **|S| ≤ 2** (within the classical limit).

**Explanation:**
- When Eve measures, she collapses the quantum superposition
- The entanglement is destroyed - Alice and Bob no longer share entangled pairs
- Instead, they effectively have classical, local correlations
- Classical systems are bounded by Bell's inequality: |S| ≤ 2
- The CHSH test would show values around |S| ≈ 2 or lower, signaling that the quantum channel has been compromised

### Question 2

Alice and Bob run the E91 protocol and measure **S = 2.1**.  
Should they use the key?

A) Yes — any $S > 2$ is considered secure  
B) No — it is too close to the classical limit  
C) Yes — but only if they discard half of the key

**SOLUTION:**

**Correct:** B

**No, they should NOT use the key.**

**Reasoning:**
- The classical limit is |S| = 2, and S = 2.1 is only slightly above this
- An ideal quantum system should give |S| ≈ 2√2 ≈ 2.83
- A value of 2.1 suggests possible eavesdropping activity (partial interception)
- It could also indicate significant channel noise
- Either way, the key's security cannot be guaranteed
- In practice, E91 implementations set a threshold well above 2 (e.g., 2.5 or higher) to account for statistical fluctuations while still detecting attacks

### Question 3

Why does the E91 protocol use a combination of **same** and **different** measurement bases?

A) Same bases for key generation; different bases allow the CHSH Bell test  
B) Same bases are required for both tasks  
C) Different bases only increase the key length

**SOLUTION:**

**Correct:** A

**Why different bases:**
- Key generation needs **perfectly correlated/anti-correlated** results → same basis (90°-90° in our case)
- The CHSH test needs **non-aligned bases** to detect quantum correlations that violate Bell inequality
- The CHSH formula uses specific basis combinations: (0°,45°), (0°,135°), (90°,45°), (90°,135°)

**If they used the same bases for both:**
- With matching bases: 100% correlation but no CHSH test possible (all terms would be the same)
- With CHSH bases only: Could do Bell test but key bits wouldn't perfectly match
- The separation is necessary to serve both purposes: security verification AND key generation

### Question 4

How does the E91 protocol detect eavesdropping?

A) By publicly comparing a subset of key bits  
B) By checking if the CHSH test violates Bell's inequality ($|S|>2$)  
C) By verifying that the encrypted message remains readable

**SOLUTION:**

**Correct:** B

**E91 Detection:**
- Uses the **CHSH Bell inequality test**
- Tests whether the correlations violate Bell's inequality (|S| > 2)
- Eavesdropping destroys entanglement, reducing S to classical levels
- Security is based on fundamental physics (entanglement properties)
